### Data set 1 - Calorie burning


Cleaning data:
1. Row 22 has missing "Date" - these data are invalid, remove this row
2. Rows 18 and 28 have missing "Calories" - data in rows is invalid, remove them
3. Row 26 had incorrect date format in column "Date" - update it to valid format
4. Convert Date column to Datetime format as they are strings. It will give you possibility to sort by date

````python
from datetime import datetime
import pandas as pd
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/Data%20Cleaning%20Data%20Sets/dirtydata.csv"
calories_df= pd.read_csv (url)

#1. Row 22 has missing "Date" - these data are invalid, remove this row
cleaned_df= calories_df.dropna(subset= ["Date"], axis=0 )

cleaned_df.shape

#2. Rows 18 and 28 have missing "Calories" - data in rows is invalid, remove them
cleaned_df = cleaned_df.dropna(subset= ["Calories"], axis=0 )
print(cleaned_df)


#3. Row 26 had incorrect date format in column "Date" - update it to valid format
date= cleaned_df['Date'].loc[cleaned_df.index[24]]
date_format= '%Y%m%d'
print (date)

formatted_date= datetime.strptime (date, date_format)
print (formatted_date)

changed_format = formatted_date.strftime('%Y/%m/%d')
print (changed_format)

#insert converted date back to dataframe
cleaned_df.at [26, 'Date'] = formatted_date
cleaned_df

#4. convert Date to datetime format
cleaned_df['Date']= pd.to_datetime(cleaned_df['Date'])
print (cleaned_df)
````

### Data set 2 - Presidential election voting
---



Clean data in dataframe presDirty:
1. Column "national_voteshare_3rd_lo" has a values Null, remove it as obsolete
2. Columns 'national_turnout', 'national_turnout_hi, and 'national_turnout_lo' have NA values, remove those values
3. There are duplicated rows, for example rows 1 ans 2. Drop duplicates
4. Some rows in 'candidate_inc' are empty, but should have value 'Trump'. Fill those rows with value.


````python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/Data%20Cleaning%20Data%20Sets/presDirty.csv"
presDirty_df= pd.read_csv (url)
presDirty_df.head (30)

#drop Column "national_voteshare_3rd_lo"
cleaned_drop3rd_df = presDirty_df.drop (["national_voteshare_3rd_lo"], axis=1)
cleaned_drop3rd_df

#Columns 'national_turnout', 'national_turnout_hi, and 'national_turnout_lo' have NA values, remove those values
cleaned_presDirty_df = cleaned_drop3rd_df.dropna (subset=["national_turnout","national_turnout_lo", "national_turnout_hi"],axis=0)

print (cleaned_presDirty_df.shape)

#drop duplicates
presDirty_df_no_dupl = cleaned_presDirty_df.drop_duplicates ()
print (presDirty_df_no_dupl.shape)

#def fill_empty_candidate (presDirty_df_no_dupl):
presDirty_df_no_dupl = presDirty_df_no_dupl.replace(r'^\s*$', np.nan, regex=True)
presDirty_df_no_dupl['candidate_inc'].fillna('Trump',inplace=True)
print (presDirty_df_no_dupl)
````
